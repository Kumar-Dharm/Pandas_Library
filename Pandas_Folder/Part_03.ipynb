### Pandas Lec-04: Arithmetic Operations in Pandas

```python
import pandas as pd
```

```python
# Arithmetic operation -Series data structures
l=[1,2,3,4,5]
idx=['a','b','c','d','e']
sr=pd.Series(l,idx)
print(sr)
sr1=sr+sr+5                      # horizontal operations
print(sr1)                                                 #   ---ask to perform vertical operations--
```

```python
# Arithmetic operation -DataFrame data structures
d={'a':[1,2,3,4,5],'b':[9,8,7,6,5],'c':[3,4,5,6,7]}
idx=[1,2,3,4,5]
df=pd.DataFrame(d,idx,columns=['a','b'])                   # we can access more than two col using columns
print(df)

df['add5']=df['a']+5
df['sub3']=df['a']-3
df['mul5']=df['a']*5
df['div5']=df['a']/5
df['mod3']=df['a']%3
df['floor3']=df['a']//3
df['power5']=df['a']**5

df['add']=df['a']+df['b']
df['sub']=df['a']-df['b']
df['mul']=df['a']*df['b']
df['div']=df['a']/df['b']
df['mod']=df['a']%df['b']
df['floor']=df['a']//df['b']
df['power']=df['a']**df['b']
df
```

### Pandas Lec-05: Logical Operators in Pandas

```python
# Logical operators in pandas -DataFrame data structures
d={'a':[10,20,30,40,50],'b':[50,40,30,20,10]}
idx=[1,2,3,4,5]
df=pd.DataFrame(d,idx)
print(df)

df['x1']=df['a']<df['b']
df['x2']=df['a']>df['b']
df['x3']=df['a']<=df['b']
df['x4']=df['a']>=df['b']
df['x5']=df['a']==df['b']
df['x6']=df['a']%5==0
df['x7']=(df['b']>=40) & (df['b']%10==0)
df['x8']=(df['b']>=20) | (df['a']<=40)
df
```
